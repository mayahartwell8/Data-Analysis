```python
#chatgpt helped
#most common shape per top 5 country
import pandas as pd

# Create a DataFrame with UFO sightings data (replace this with your actual UFO sightings data)
data = {
    'country': ['us','gb','ca','au','de','us','gb','ca','au','de','us','gb','ca','au','de','us','gb','ca','au','de','us','gb','ca','au','de','us','gb','ca','au'],
    'shape': ['cylinder', 'light', 'circle', 'sphere', 'disk', 'fireball',
       'unknown', 'oval', 'other', 'cigar', 'rectangle', 'chevron',
       'triangle', 'formation', 'delta', 'changing', 'egg', 'flash',
       'diamond', 'cross', 'teardrop', 'cone', 'pyramid', 'round',
       'crescent', 'flare', 'hexagon', 'dome', 'changed'],
}

ufo_sightings_df = pd.DataFrame(data)

# Define a list of countries
countries =['us','gb','ca','au','de']

# Iterate through countries and find the most common shape for each
for country in countries:
    country_df = ufo_sightings_df[ufo_sightings_df['country'] == country]
    most_common_shape = country_df['shape'].mode().values[0]
    print(f'Most common UFO shape in {country}: {most_common_shape}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm  # Add this import statement

# Create a DataFrame with UFO sightings data (replace this with your actual UFO sightings data)
data = {
    'state': ['wa', 'az', 'ca', 'il', 'oh', 'mi', 'pa', 'fl', 'ny', 'tx'],
    'UFO Sightings': [4563, 2940, 10450, 2874, 2642, 2254, 2783, 4598, 3511, 4050]  # Replace with actual UFO sightings counts
}

ufo_sightings_df = pd.DataFrame(data)

# Create a DataFrame with population data for the specified states
population_data = {
    'state': ['wa', 'az', 'ca', 'il', 'oh', 'mi', 'pa', 'fl', 'ny', 'tx'],
    'population': [7693612, 7378494, 39538223, 12659682, 11799448, 10077331, 12812508, 21538187, 20201249, 29145505]  # Population data is approximate
}

population_df = pd.DataFrame(population_data)

# Merge the UFO sightings and population DataFrames on the 'state' column
merged_data = pd.merge(ufo_sightings_df, population_df, on='state')

# Calculate UFO sightings per capita by dividing UFO sightings by population
merged_data['UFO Sightings per Capita'] = merged_data['UFO Sightings'] / merged_data['population']

# Print the resulting DataFrame with UFO sightings per capita
print(merged_data)

# Create a bar graph of UFO sightings per capita with a gradient
plt.figure(figsize=(10, 6))
colors = cm.viridis(merged_data['UFO Sightings per Capita'] / max(merged_data['UFO Sightings per Capita']))  # Use a colormap for the gradient
bars = plt.bar(merged_data['state'], merged_data['UFO Sightings per Capita'], color=colors)
plt.xlabel('State')
plt.ylabel('UFO Sightings per Capita')
plt.title('UFO Sightings per Capita by State')
plt.xticks(rotation=45)

# Show the bar graph
plt.tight_layout()
plt.show()
```
