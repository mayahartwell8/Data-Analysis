```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df' with columns: generation, hp, atk, def, spatk, spdef, speed

# Group the data by generation and calculate the mean stats for each generation
generation_stats = df.groupby('generation').mean()

# Define the generations as x-axis labels
generations = generation_stats.index

# Define the stats you want to compare
stats = ['hp', 'atk', 'def', 'speed']

# Create a line chart for each stat
plt.figure(figsize=(12, 6))

for stat in stats:
    plt.plot(generations, generation_stats[stat], label=stat)

# Customize the plot
plt.title('Pokmon Stat Comparison by Generation')
plt.xlabel('Generation')
plt.ylabel('Stat Value')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'df' with 'type1' and 'type2' columns

# Calculate the count of each Type 1
type1_counts = df['type1'].value_counts()

# Calculate the count of each Type 2
type2_counts = df['type2'].value_counts()

# Combine counts for Type 1 and Type 2 into a single DataFrame
type_counts = pd.DataFrame({'Type 1': type1_counts, 'Type 2': type2_counts}).fillna(0)

# Create a stacked bar chart
ax = type_counts.plot(kind='bar', stacked=True, figsize=(12, 6))

# Set plot labels and title
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Stacked Bar Chart of Pokmon Type 1 and Type 2')

# Show the plot
plt.legend(title='Type')
plt.tight_layout()
plt.show()
```
